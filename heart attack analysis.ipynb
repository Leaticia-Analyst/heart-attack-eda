## Project objective
The Objective of this project is to perform Exploratory Data Analysis (EDA) on a heart attack dataset in order to understand the distribution and relationships between key cardiovascular risk factors.

This analysis focuses on patient characteristics such as:

- Age  
- Blood Pressure  
- Cholesterol Level  
- Diabetes Status  
- Smoking Status  
- Chest Pain Type  
- Treatment Type  

The goal is to identify patterns, trends, and potential correlations that may help in understanding factors associated with heart disease.

```python
## Dataset Description

The dataset contains 1000 patient records with 8 variables.

Variables included in the dataset:

- Gender: Male or Female  
- Age: Age of the patient  
- Blood Pressure (mmHg): Patient’s blood pressure level  
- Cholesterol (mg/dL): Cholesterol measurement  
- Has Diabetes: Whether the patient has diabetes (Yes/No)  
- Smoking Status: Never, Former, or Current smoker  
- Chest Pain Type: Type of chest pain reported  
- Treatment: Type of treatment provided  

The dataset appears to contain clean and structured medical information suitable for exploratory analysis.
```

## Correlation Analysis

A correlation heatmap was generated to examine relationships between numerical variables.

Observations:

- Age shows almost no correlation with cholesterol or blood pressure.
- Blood pressure and cholesterol also show very weak correlation.
- All correlation values are close to zero.

This indicates that the numerical variables do not have strong linear relationships and may independently contribute to cardiovascular risk.

```python
correlation = heart.select_dtypes(include='number').corr()
```

```python
sns.heatmap(correlation, xticklabels=correlation.columns, yticklabels=correlation.columns,annot=True)
```

```python
sns.pairplot(heart)
```

```python
sns.relplot(x= 'Cholesterol (mg/dL)', y= 'Age', hue= 'Gender', data=heart)
```

```python
sns.histplot(heart['Gender'])
```

```python
sns.boxplot(x= 'Cholesterol (mg/dL)', data= heart)
```
